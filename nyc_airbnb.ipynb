```python
columns_to_drop_ver_1 = ["listing_url", "scrape_id", "last_scraped", "source", "description", "neighborhood_overview", "picture_url", "host_url", "host_about", "host_thumbnail_url", "host_picture_url", "calendar_last_scraped", "first_review", "license", "reviews_per_month", "host_since", "host_location", "neighbourhood", "minimum_nights", "maximum_nights", "minimum_minimum_nights", "maximum_minimum_nights", "minimum_maximum_nights", "maximum_maximum_nights", "minimum_nights_avg_ntm", "maximum_nights_avg_ntm", "calendar_updated", "availability_30", "availability_60", "availability_90", "availability_365", "number_of_reviews_ltm", "number_of_reviews_l30d", "property_type", "host_has_profile_pic", "host_identity_verified", "calculated_host_listings_count", "calculated_host_listings_count_entire_homes", "calculated_host_listings_count_private_rooms", "calculated_host_listings_count_shared_rooms", "bathrooms", "host_listings_count", "host_total_listings_count", "host_verifications"]

df_lean_ver_1 =  listing_csv_gz_df.drop(columns = columns_to_drop_ver_1)
df_lean_ver_1
```

```python
# Remove the '%' sign and ',' character, and convert the column to float
df_lean_ver_1['host_response_rate'] = df_lean_ver_1['host_response_rate'].str.replace('%', '').astype(float)
```

```python
# Remove the '%' sign and ',' character, and convert the column to float
df_lean_ver_1['host_acceptance_rate'] = df_lean_ver_1['host_acceptance_rate'].str.replace('%', '').astype(float)
```

```python
df_lean_ver_1['price'] = df_lean_ver_1['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
df_lean_ver_1.info()
```

```python
df_lean_ver_1
```

```python
# Convert the 'last_review' column to datetime
df_lean_ver_1['last_review'] = pd.to_datetime(df_lean_ver_1['last_review'])
```

```python
df_lean_ver_1.info()
```

```python
# has availability is true

condition = df_lean_ver_1['has_availability'] != "t" 
has_availability_is_true = df_lean_ver_1[~condition]

has_availability_is_true
```

```python
has_availability_is_true.info()
```

```python
# Remove rows where number_of_reviews == 0
condition = has_availability_is_true['number_of_reviews'] == 0  
drop_reviews_of_zero = has_availability_is_true[~condition]

drop_reviews_of_zero
```

```python
# year, month, date
# Calculate the date threshold for filtering
threshold_date = datetime(2023, 2, 1)  # Feb 1, 2023 threshold date

# Filter rows where 'date' is before the threshold date
condition = drop_reviews_of_zero['last_review'] < threshold_date
df_filtered = drop_reviews_of_zero[~condition]

df_filtered
```
